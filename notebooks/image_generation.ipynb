```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import arabic_reshaper
from bidi.algorithm import get_display

text = "أعزب"
reshaped_text = arabic_reshaper.reshape(text)
bidi_text = get_display(reshaped_text)
font_path = r"D:\Users\HP\Downloads\simplified-arabic-bold\Simplified Arabic Bold\Simplified Arabic Bold.ttf"
prop = fm.FontProperties(fname=font_path, size=80)
fig, ax = plt.subplots(figsize=(8, 6))
ax.axis("off")
plt.text(0.5, 0.5, bidi_text, fontproperties=prop,
         ha="center", va="center")

plt.show()
```
